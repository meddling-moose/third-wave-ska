```python
## Create a new DataFrame to better make timeline visualizations with
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import nltk
from nltk.sentiment import SentimentIntensityAnalyzer
from pathlib import Path

DATA_DIR = Path("../data")
df = pd.read_csv(DATA_DIR / "processed/artist_counts_locations_years.csv")

df.shape
```

```python
df.head()
```

```python
plt.figure(figsize=(10, 6))
df.plot(kind="bar",x="artist", y="count")
plt.xlabel("Artist")
plt.ylabel("Number of Songs")
plt.title("Number of Songs Per Artist")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```
